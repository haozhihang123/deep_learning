```python
import torch
from torch.autograd import Variable
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
x = torch.unsqueeze(torch.linspace(-1,1,100),dim=1)
y = x.pow(3) + 0.2*torch.rand(x.size())

x,y = Variable(x),Variable(y)

plt.scatter(x.data,y.data)
plt.show()
```

```python
# net = torch.nn.Sequential(
#     # 首先通过其完成从输入层到隐藏层的线性变换
#     torch.nn.Linear(1,10),
#     # 经过激活函数
#     torch.nn.ReLU(),
#     # 接着完成从隐藏层1到隐藏层2的线性变换
#     torch.nn.Linear(10,15),
#     # 经过激活函数
#     torch.nn.ReLU(),
#     # 最后完成从隐藏层2到输出层的线性变换
#     torch.nn.Linear(15,1)
# )

class net1(torch.nn.Module):
    def __init__(self):
        super(net1, self).__init__()
        self.linear1 = torch.nn.Linear(1,10)
        self.linear2 = torch.nn.Linear(10,15)
        self.linera_out = torch.nn.Linear(15,1)
    def forward(self, x):
        out = F.relu(self.linear1(x))
        out = F.relu(self.linear2(out))
        out = self.linera_out(out)
        return out
net = net1()
print(net1)
```

```python
optimizer = torch.optim.SGD(net.parameters(), lr = 1e-1)
loss_func = torch.nn.MSELoss()

for t in range(5000):
    prediction = net(x)
    loss = loss_func(prediction, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if t%100 == 0:
        plt.cla()
        plt.scatter(x.data.numpy(),y.data.numpy())
        plt.plot(x.data.numpy(),prediction.data.numpy(),'r-',lw=3)
        plt.text(0.5,0,'loss=%.4f'%loss.data.item(),fontdict={'size':20,'color':'red'})
        plt.pause(0.1)
        print(loss)
    
# print(prediction)
# print(y)
```
